```python
# Imports
import pickle
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
# Read in data
with open("df_plot.pkl", "rb") as f:
    df_plot = pickle.load(f)
```

```python
df_plot[(df_plot["nn"] == "DRN") & (df_plot["metric"] == "w")]
```

```python
nn_vec = ["drn", "bqn"]
score_labels = {
    "crps": "CRPS",
    "crpss": "CRPSS in %",
    "me": "Bias",
    "lgt": "PI length",
    "cov": "PI coverage in %",
    "a": "Intercept",
    "w": "Rel. weight diff. in %",
}

fig, axes = plt.subplots(len(score_labels), len(nn_vec), figsize=(15, 20))

for i, temp_nn in enumerate([x.upper() for x in nn_vec]):
    for j, metric in enumerate(score_labels.keys()):
        sns.lineplot(
            ax=axes[j, i],
            data=df_plot[(df_plot["nn"] == temp_nn) & (df_plot["metric"] == metric)],
            x="n_ens",
            y="score",
            hue="agg",
            
        )
        axes[j, i].set_ylabel(score_labels[metric])
```
